### Saving Models

```python

import os
import pickle
import numpy as np
import yaml
from statsforecast.models import AutoARIMA
from data_formatter.base import DataFormatter  # make sure path is correct

def train_and_save_arima_models(dataset_name: str, save_dir="./best_arima_models"):
    # === Load Dataset ===
    config_path = f"./config/{dataset_name}.yaml"
    with open(config_path, "r") as f:
        config = yaml.safe_load(f)

    config['scaling_params']['scaler'] = 'MinMaxScaler'
    formatter = DataFormatter(config)

    in_len = formatter.params['max_length_input']
    out_len = formatter.params['length_pred']
    target_col = formatter.get_column('target')
    group_col = formatter.get_column('sid')

    # Use full training data
    train_data = formatter.train_data

    # === Track Best Model ===
    best_mae = float("inf")
    best_model = None
    best_group = None

    # === Train Models for Each Group ===
    for group, data in train_data.groupby(group_col):
        series = data[target_col].iloc[:in_len].values.flatten()

        model = AutoARIMA(
            start_p=0, max_p=10,
            start_q=0, max_q=10,
            start_P=0, max_P=10,
            start_Q=0, max_Q=10,
            allowdrift=True,
            allowmean=True
        )

        model.fit(series)

        # Evaluate on next window
        valid_start = in_len
        valid_end = valid_start + out_len
        if valid_end > len(data):
            continue  # not enough data

        input_series = data[target_col].iloc[valid_start - in_len:valid_start].values.flatten()
        true_vals = data[target_col].iloc[valid_start:valid_end].values.flatten()

        prediction = model.forecast(input_series, h=out_len)['mean']

        mae = np.mean(np.abs(true_vals - prediction))

        if mae < best_mae:
            best_mae = mae
            best_model = model
            best_group = group

    print(f"✅ Best group: {best_group} with MAE: {best_mae:.4f}")

    # === Save Best Model ===
    os.makedirs(save_dir, exist_ok=True)
    save_path = os.path.join(save_dir, f"best_arima_model_{dataset_name}.pkl")

    with open(save_path, "wb") as f:
        pickle.dump(best_model, f)
   
    print(f"✅ Best model {dataset_name} saved from group {best_group} with MAE={best_mae:.4f}")

def load_arima_model(model_path):
    with open(model_path, 'rb') as f:
        return pickle.load(f)
```

```python
import re
import pandas as pd
import matplotlib.pyplot as plt

def extract_metrics(file_path):
    with open(file_path, 'r') as f:
        content = f.read()

    # Use regex to find the last occurrence of overall statistics
    id_mean = re.findall(r'ID mean of \(MSE, MAE\): \[([\d.]+), ([\d.]+)\]', content)
    ood_mean = re.findall(r'OOD mean of \(MSE, MAE\): \[([\d.]+), ([\d.]+)\]', content)
    id_median = re.findall(r'ID median of \(MSE, MAE\): \[([\d.]+), ([\d.]+)\]', content)
    ood_median = re.findall(r'OOD median of \(MSE, MAE\): \[([\d.]+), ([\d.]+)\]', content)

    return {
        'ID_MSE_mean': float(id_mean[-1][0]) if id_mean else None,
        'ID_MAE_mean': float(id_mean[-1][1]) if id_mean else None,
        'OOD_MSE_mean': float(ood_mean[-1][0]) if ood_mean else None,
        'OOD_MAE_mean': float(ood_mean[-1][1]) if ood_mean else None,
        'ID_MSE_median': float(id_median[-1][0]) if id_median else None,
        'ID_MAE_median': float(id_median[-1][1]) if id_median else None,
        'OOD_MSE_median': float(ood_median[-1][0]) if ood_median else None,
        'OOD_MAE_median': float(ood_median[-1][1]) if ood_median else None,
    }

def compare_metrics(file1, file2):
    metrics1 = extract_metrics(file1)
    metrics2 = extract_metrics(file2)

    df = pd.DataFrame({
        'Metric': metrics1.keys(),
        'Original': list(metrics1.values()),
        'New': [metrics2[k] for k in metrics1.keys()]
    })

    df['Difference'] = df['Original'] - df['New']
    df['% Change'] = 100 * df['Difference'] / df['Original']
    return df

def plot_histograms(df):
    metrics = df['Metric']
    original = df['Original']
    new = df['New']

    x = range(len(metrics))
    width = 0.35

    plt.figure(figsize=(12, 6))
    plt.bar(x, original, width=width, label='Original', color='skyblue')
    plt.bar([i + width for i in x], new, width=width, label='New', color='lightgreen')
    plt.xticks([i + width / 2 for i in x], metrics, rotation=45, ha='right')
    plt.ylabel('Metric Value')
    plt.title('ARIMA Metric Comparison')
    plt.legend()
    plt.tight_layout()
    plt.show()

# === Usage ===
file1 = '/kaggle/input/dataset-glucobench/output/arima_hall.txt'
file2 = '/kaggle/working/output/arima_hall.txt'

comparison_df = compare_metrics(file1, file2)
print(comparison_df.to_string(index=False))

# === Plotting ===
plot_histograms(comparison_df)
```

# Train usinfied AutoArima model with five datasets

```python
import os
import pickle
import numpy as np
import yaml
from statsforecast.models import AutoARIMA
from data_formatter.base import DataFormatter  # adjust path if needed

def train_unified_arima_on_datasets(dataset_names, save_dir="./best_arima_models", model_name="unified_arima"):
    combined_series = []

    for dataset_name in dataset_names:
        # Load dataset config
        config_path = f"./config/{dataset_name}.yaml"
        if not os.path.exists(config_path):
            print(f"⚠️ Config not found for {dataset_name}, skipping.")
            continue

        with open(config_path, "r") as f:
            config = yaml.safe_load(f)

        config['scaling_params']['scaler'] = 'MinMaxScaler'
        formatter = DataFormatter(config)

        # Extract parameters
        target_col = formatter.get_column('target')
        group_col = formatter.get_column('sid')

        # Get train data from all groups
        train_data = formatter.train_data
        for group, data in train_data.groupby(group_col):
            series = data[target_col].values.flatten()
            if len(series) > 0:
                combined_series.append(series)

    # Flatten all series into one big time series
    full_series = np.concatenate(combined_series)

    # Sanity check
    if len(full_series) < 100:
        print("❌ Combined series too short for ARIMA. Aborting.")
        return None

    print(f"✅ Combined training series length: {len(full_series)}")

    # Train unified ARIMA model
    model = AutoARIMA(
        start_p=0, max_p=10,
        start_q=0, max_q=10,
        start_P=0, max_P=10,
        start_Q=0, max_Q=10,
        allowdrift=True,
        allowmean=True
    )
    model.fit(full_series)

    # Save model
    os.makedirs(save_dir, exist_ok=True)
    model_path = os.path.join(save_dir, f"{model_name}.pkl")
    with open(model_path, "wb") as f:
        pickle.dump(model, f)

    print(f"✅ Unified ARIMA model trained and saved to: {model_path}")
    return model
```

```python
datasets = ["iglu", "colas", "dubosson", "weinstock", "hall"]
train_unified_arima_on_datasets(datasets, model_name="arima_all_5")
```
